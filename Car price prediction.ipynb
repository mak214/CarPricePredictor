## Model

```python
X = car.drop(columns="Price")
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
score = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    score.append(r2_score(y_test,y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=np.argmax(score))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
best_score = -np.inf
best_X_train, best_X_test, best_y_train, best_y_test = None, None, None, None

for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(X_train, y_train)
    
    y_pred = pipe.predict(X_test)
    current_score = r2_score(y_test, y_pred)
    
    if current_score > best_score:
        best_score = current_score
        best_X_train, best_X_test, best_y_train, best_y_test = X_train, X_test, y_train, y_test

# Now train on the best split:
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(best_X_train, best_y_train)
y_pred = pipe.predict(best_X_test)

print(r2_score(best_y_test, y_pred))  # Will match your best score!
```

```python
def getMaxPerformanceSum(performanceScores, activationStatus, k):
    n = len(performanceScores)
    
    # Step 1: Create a modified array that only includes active servers' performance scores
    activePerformance = [performanceScores[i] if activationStatus[i] == 1 else 0 for i in range(n)]
    
    # Step 2: Use sliding window to calculate the maximum sum of at most k consecutive active servers
    max_sum = 0
    current_sum = 0
    window_start = 0
    
    for window_end in range(n):
        current_sum += activePerformance[window_end]
        
        # If the window size exceeds k, slide the window by removing the leftmost element
        if window_end - window_start + 1 > k:
            current_sum -= activePerformance[window_start]
            window_start += 1
        
        # Track the maximum sum found so far
        max_sum = max(max_sum, current_sum)
    
    return max_sum

# Example usage:
n = 4
k = 2
performanceScores = [7, 4, 3, 5]
activationStatus = [1, 0, 0, 0]

print(getMaxPerformanceSum(performanceScores, activationStatus, k))
```
